# Applying `Gradient Descent`

```python
# iteration - 0
# known slope as m = 78.35 and b = 0
m = 78.35
b = 0
y_pred0 = m * X.ravel() + b
y_pred0
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red', label='OLS')
plt.plot(X, y_pred0, color='#A3E4D7', label='b = 0')
plt.legend()
plt.plot()
```

```python
# iteration - 1
m = -2 * np.sum(y - (m * X.ravel()) - b)
m
```

```python
lr = 0.1
b = b - (lr*m)
b
```

```python
y_pred1 = 78.35 * X.ravel() + b
y_pred1
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red', label='OLS')
plt.plot(X, y_pred0, color='#A3E4D7', label='b = 0')
plt.plot(X, y_pred1, color='#A3E4D7', label='b = 32')
plt.legend()
plt.plot()
```

```python
# iteration - 2
m = -2 * np.sum(y - (78.35 * X.ravel()) - b)
m
```

```python
b = b - (lr*m)
b
```

```python
y_pred2 = 78.35 * X.ravel() + b
y_pred2
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red', label='OLS')
plt.plot(X, y_pred0, color='#A3E4D7', label='b = 0')
plt.plot(X, y_pred1, color='#A2E1F7', label='b = 20')
plt.plot(X, y_pred2, color='#000', label='b = 25.11')
plt.legend()
plt.plot()
```

```python
# iteration - 3
m = -2 * np.sum(y - (78.35*X.ravel()) - b)
m
```

```python
b = b - (lr * m)
b
```

```python
y_pred3 = 78.35 * X.ravel() + b
y_pred3
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red', label='OLS')
plt.plot(X, y_pred0, color='#A3E4D7', label='b = 0')
plt.plot(X, y_pred1, color='#d65db1', label='b = 20')
plt.plot(X, y_pred2, color='#000', label='b = 25.11')
plt.plot(X, y_pred3, color='purple', label='b = 25.11')
plt.legend()
plt.plot()
```

```python
fig = plt.figure(figsize=(15, 7))
# applying using loop
m = 78.35
b = -100
lr = 0.12
epochs = 5
for i in range(epochs):
  slope_loss = -2 * np.sum(y - (m*X.ravel()) - b)
  b = b - (lr*slope_loss)
  print(i, " b: ", b)

  y_pred = m * X.ravel() + b

  plt.plot(X, y_pred, label=f'At b = {b}')
plt.legend()
plt.scatter(X, y)
plt.show()
```

# Creating own class for Gradint Descent

```python
X, y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=20)
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
lr.intercept_
```

```python
m = 20.82
```

```python
class GDRegressor:
  def __init__(self, learning_rate, epochs):
    self.m = 20.82
    self.b = -120
    self.lr = learning_rate
    self.epochs = epochs

  def fit(self, X, y):
    for i in range(self.epochs):
      slope_loss = -2 * np.sum(y - (self.m*X.ravel()) - self.b)
      self.b = self.b - (self.lr * slope_loss)
    print(self.b)
```

```python
gd = GDRegressor(0.001, 50)
```

```python
gd.fit(X, y)
```
